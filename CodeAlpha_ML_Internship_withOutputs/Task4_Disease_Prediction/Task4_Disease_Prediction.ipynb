```
# Load dataset.csv and train RandomForest classifier
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier

df = pd.read_csv('dataset.csv')
X = df.drop(columns='disease').values; y = df['disease'].values
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,stratify=y,random_state=42)
scaler = StandardScaler().fit(X_train)
X_train_s = scaler.transform(X_train); X_test_s = scaler.transform(X_test)
rf = RandomForestClassifier(n_estimators=200, random_state=42)
rf.fit(X_train_s, y_train)
y_pred = rf.predict(X_test_s)
from sklearn.metrics import classification_report, roc_auc_score
print(classification_report(y_test, y_pred))
print('ROC-AUC:', roc_auc_score(y_test, rf.predict_proba(X_test_s)[:,1]))
```
